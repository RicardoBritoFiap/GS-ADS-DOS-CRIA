```python
import PyPDF2

def extrair_texto_de_pdf(caminho_do_pdf):
    # Abrindo o arquivo PDF em modo binário
    with open(caminho_do_pdf, 'rb') as arquivo:
        leitor_pdf = PyPDF2.PdfReader(arquivo)
        num_paginas = len(leitor_pdf.pages)

        # Lendo cada página e extraindo o texto
        texto_completo = ''
        for pagina in range(num_paginas):
            texto_pagina = leitor_pdf.pages[pagina].extract_text()
            texto_completo += texto_pagina

    return texto_completo

# Exemplo de uso
caminho_do_pdf = r'C:\Users\pytho\Downloads\Hemograma_Completo.pdf'  # Substitua pelo caminho do seu arquivo PDF
texto_extraido = extrair_texto_de_pdf(caminho_do_pdf)
# print(texto_extraido)
```

```python
import re

def extrair_dados_hemograma(texto):
    dados = {}

    # Extrair os valores do hemograma
    padroes = {
        'Eritrócitos': r'Eritrócitos[^\d]+(\d+,\d+)',
        'Hemoglobina': r'Hemoglobina[^\d]+(\d+,\d+)',
        'Hematócrito': r'Hematócrito[^\d]+(\d+,\d+)',
        'V.C.M.': r'V.C.M.[^\d]+(\d+,\d+)',
        'H.C.M.': r'H.C.M.[^\d]+(\d+,\d+)',
        'C.H.C.M.': r'C.H.C.M.[^\d]+(\d+,\d+)',
        'RDW (Indice de Ansiocitose)': r'RDW \(Indice de Ansiocitose\)[^\d]+(\d+,\d+)',
        'Leucócitos': r'Leucócitos[^\d]+(\d+,\d+)',
        'Mielócitos': r'Mielócitos[^\d]+(\d+,\d+)',
        'Metamielócitos': r'Metamielócitos[^\d]+(\d+,\d+)',
        'Bastonetes': r'Bastonetes[^\d]+(\d+,\d+)',
        'Segmentados': r'Segmentados[^\d]+(\d+,\d+)',
        'Eosinófilos': r'Eosinófilos[^\d]+(\d+,\d+)',
        'Basófilos': r'Basófilos[^\d]+(\d+,\d+)',
        'Linfócitos Típicos': r'Linfócitos Típicos[^\d]+(\d+,\d+)',
        'Linfócitos Atípicos.': r'Linfócitos Atípicos\.[^\d]+(\d+,\d+)',
        'Monócitos': r'Monócitos[^\d]+(\d+,\d+)',
        'Plaquetas': r'Plaquetas[^\d]+(\d+,\d+)',
    }

    for chave, padrao in padroes.items():
        match = re.search(padrao, texto)
        if match:
            dados[chave] = match.group(1).replace(',', '.')

    return dados

# Exemplo de uso
texto_extrai = texto_extraido
dados_hemograma = extrair_dados_hemograma(texto_extrai)
# print(dados_hemograma)
```
